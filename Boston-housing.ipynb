```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path = 'boston_housing.npz')
```

Let's look into boston_housing dataset, a built-in dataset for keras.

```python
print(x_train.shape)
print(x_test.shape)
```

404 samples with 13 parameters / 102 samples for testing

Double-check the format of the data.

어떤 trait이 의미있는지 보려면, 그냥 뭘 띄워보는게 낫다.

```python
plt.plot(x_train[:,0], y_train, '.') 
#이런 식으로 경향성을 확인 가능. 0일 경우 범죄율 data 띄워본거. 분명히 관계가 있긴 있는데, linear하지는 않다.
```

```python
plt.plot(x_train[:,1], y_train, '.') 
#이런 식으로 경향성을 확인 가능. 이건 경향이 더 모호...하긴 한데. 0 퍼센트에 걸려있는 경향성이 좀 크지.
```

서브플롯 활용해서 데이터 한번 띄워보고, 좀 경향성을 이해해보자.

```python
plt.plot()
boston_housing = ['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptrate','black','lstat']
for i in range(13):
    plt.subplot(3,5,i+1) # 3x5의 subplot 공간의 i+1번째에 플랏팅할것.
    plt.plot(x_train[:,i], y_train,'.')
    plt.title(boston_housing[i])
plt.show
```

```python
model = Sequential()
model.add(Dense(1, input_shape = [13])) #I/O 확정, 인풋 13개와 아웃풋 단 하나.
model.compile(optimizer = 'rmsprop', loss = 'mse')
model.summary() #quick summary of the data. bias 있으니까 parameter가 하나 더 있어보이는거는 아실테고.
```

```python
model.fit(x_train, y_train, epochs = 1000)
```

당연히 이 loss가 줄면 줄수록 좋다.

```python
model.evaluate(x_test, y_test) # part of dataset for training, another for testing. sure thing.
```

```python
print(model.input_shape)
print(model.output_shpae) 
# we only have one output in sequential model, so this inflicts an error.
```

We can actually make predictions by using model.predict with utilizing parameters we trained and putting a random data into it.

```python
# INPUT::
# 1. crim: per capita crime rate by town.
# 2. zn: proportion of residential land zoned for lots over 25,000 sq.ft.
# 3. indus: proportion of non-retail business acres per town.
# 4. chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
# 5. nox: nitrogen oxides concentration (parts per 10 million).
# 6. rm: average number of rooms per dwelling.
# 7. age: proportion of owner-occupied units built prior to 1940.
# 8. dis: weighted mean of distances to five Boston employment centres.
# 9. rad: index of accessibility to radial highways.
# 10. tax: full-value property-tax rate per $10,000. 세제가 달라서, 집값과 세금이 비례하지 않음.
# 11. ptratio: pupil-teacher ratio by town.
# 12. black: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
# 13. lstat: lower status of the population (percent).

# TARGET::
# medv: median value of owner-occupied homes in $1000s.

model.predict(np.array([0, 1, 50, 1, 0, 5, .5, 10, 1000, 100, 10, 100, 1]).reshape(1,13))
```

```python
# with hidden layers.
model = Sequential()
model.add(Dense(10, input_shape = [13],activation = 'relu')) #I/O 확정, 인풋 13개와 아웃풋 단 하나. 10개짜리 은닉층이랑.
model.add(Dense(1)) # hidden layer 이후에는 인풋형태 넣어줄 필요 없다.
model.compile(optimizer = 'rmsprop', loss = 'mse')
model.summary() #quick summary of the data. bias 있으니까 parameter가 하나 더 있어보이는거는 아실테고.
```

파라미터 수야 당연히 (13+1) * 10 = 140, 마지막 층은 은닉층 노드+1 해서 파라미터 11개

bias는 매 층에 갖다붙여야하니까.

```python
model.fit(x_train, y_train, epochs = 1000)
```
