정규표현 메소드는 다 스트링에만 먹는 거였었지. 타입을 계속 쫓아가주어야 함.
<br>아래 코드들은 정규표현식 활용 + list comprehension이 전부. + print 코드에 [:10] 이건 다 redundant함..
<br> re.search(key_word, target_word), re,match(key_word, target_word), re.findall(key_word, target_word)
<br> 서치나 매치 같은거로 뱉는 내용물은 레인지랑 일치 문자열 알려주는 정규표현 객체
<br> 저 밑에 있는 if re.search ~~~ 는 만약 대상 문자열을 찾을 수 있으면 인덱스를 주니까, 그럼 if 3 같은거랑 똑같지. if에 0 빼고 다른 거 넣음 true인건 당연하고.

```python
print(wordlist[:10])
result = [w for w in wordlist if re.search('ed$', w)][:10] # ed로 끝나는 단어
print(result)
result = [w for w in wordlist if re.search('^..j..t..$', w)][:10] #.을 와일드카드라고 하네여, len = 8 이고 j,t가 지정된 위치에 있는거 검색.
print(result)
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)][:10]
print(result)
result = [w for w in wordlist if re.search('^[ah]+$', w)][:10]
print(result)

wsj = sorted(set(nltk.corpus.treebank.words())) #월스트리트 저널 데이터래 갸
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)] # \.이 .자체를 찾는 방법이었지
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)] #리얼 $를 찾기 위함. A-Z했으니 대문자만 긁히고.
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)] #0에서 9까지 4번 나오고 그게 끝인 조합.
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)] # +야 하나 이상 이야기하는거고, -는 그냥 대시임. 문장기호. 그게 끝.
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]# 알파벳 5개 이상-2~3개-최대 6개... 왜?? ㅡ 뭐 다 쓸데 있다.
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('(ed|ing)$', w)] # 괄호는 별 의미가 없음. | == or라는거 쉽게 알만 함.| 개수 늘이면 더 늘어나는거 주의.
print(wordlist[:10])
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word)) #comprehension을 쌓아뒀네.
fdist.most_common(10)
```

제법 난장판임. 다중 for 때문에 보기가 좀 복잡함. 
<br> findall 활용법 주의.
<br> re.findall(r'[aeiou]{2,}',word) 이게 어떻게 보면 리스트 내장의 극단적인 장점 예임.
<br> nltk.FreqDist(~)는 그 분포량 확인해주는 메소드.

```python
word = 'supercalifragilisticexpialidocious' #해보는게 숙제.
result = re.findall(r'[aeiou]', word) #모음 찾아라 - r은 사실 별로 차이가 없다. 애초에 이스케이프가 없는 문자열을 넣었으니까.
print(result)
result = re.findall('[aeiou](..)[aeiou]', word) #findall이 괄호친거만 뱉으니까. 모음+ XX + 모음인 형태인 부분의 XX만 긁어봐라
print(result)
result = re.findall('[^aeiou].+[^aeiou]', word) #모음이 아닌걸로 시작하고, 아무 문자나 1개 이상 연결, 이 다음에 모음이 아닌걸로 닫아라.
print(result)
result = re.findall('[^aeiou].+?[^aeiou]', word) #위에랑 같은 조건이지만, 최소매칭 전체검색!
print(result)
```

```python
word = 'bap'
result = re.findall('[^aeiou].+?[^aeiou]', word) # [ ]안에 있는 ^는 not의 의미다!!!!
print(result)
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]' 
#raw text meaning... multiple vowels as start or multiple vowels as endings or multiple consonants.

def compress(word): #defining a fuction... takes word as argument.
    pieces = re.findall(regexp, word) #find all matching strings from argument
    return ''.join(pieces) #and join them into a long string

english_udhr = nltk.corpus.udhr.words('English-Latin1') #.words해서 내놓은 리턴이니까, 스트링 덩어리 리스트임!
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

결국, compress는 텍스트 내 단어들의 어두/어미에 있는 모음만 남기고, 나머지 모음은 날린 것들을 findall로 긁고 이걸 다시 한 덩어리로 짜부라트려 주는 것.

compress가 받아오는 논항은 nltk에서 뽑아주는 코퍼스 내용을 담을 리스트. for구문으로 여기서 단어 하나씩 뽑아와서 원하는 처리를 하고, 다시 짜부라트리는 것을 반복. 마지막으로 tokenwrap이 ' '.join(target_list)한 거랑 같은 형태로 병합시켜줌.

아래 두 코드는 확인용 참고자료

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]' 
#raw text meaning... multiple vowels as start or multiple vowels as endings or multiple consonants.

def compress(word): #defining a fuction... takes word as argument.
    pieces = re.findall(regexp, word) #find all matching strings from argument
    return ''.join(pieces) #and join them into a long string
j = ['this','lists','are','a','pain','in','the','neck']
print(nltk.tokenwrap(compress(w) for w in j))
#for clarification.. how does nltk.tokenwrap work?
nltk.tokenwrap(['omg','lul','holy','cow'])
```

사실 위에서 한 거는 re.compile 써서 정규표현 객체 만드는거랑 비슷한 작업임.

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic') #gets list of words
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)] 
# rotokas_words 안에 있는 단어들 꺼내서, 특정 두 자모 조합인 syllable들을 찾고-findall이니까 리스트로 나옴 
# 그걸 각각 리스트 내장 구문을 활용해서 cvs 리스트 내에 할당.
cfd = nltk.ConditionalFreqDist(cvs) # CFD works pairwise only, 분포 확인함.
cfd.tabulate() #표로 내어놓음.
```

```python
cv_word_pairs = [(cv, w) for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
#아까 거랑 같은 상황에서 리스트에 검색된 부분, 전체 단어를 튜플로 한꺼번에 집어넣음.
cv_index = nltk.Index(cv_word_pairs)
#방금 만들어놓은 튜플들에서, 검색된 부분을 키로 - 전체 단어를 밸류로 한 유사 딕셔너리 자료형으로 병합.
#자료형 자체는 nltk.util.index로 나옴!
cv_index['po']
```
